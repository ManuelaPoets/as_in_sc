```python
# Generate total_reads_DistalLungEpithelium.csv file for SingleSplice input
# mapped mouse files concatenated, total mouse files concatenated by cat mapped*_mouse.txt, tot*_mouse.txt
# same for ERCC files 

import csv
import pandas as pd
import numpy as np
from numpy import savetxt

# ERCC reads
dir_ERCC_map='./SiSp_read_counts/mapped_ERCC.txt'
dir_ERCC_tot='./SiSp_read_counts/tot_ERCC.txt'

# mouse genome reads
dir_mouse_map='./SiSp_read_counts/mapped_mouse.txt'

# read IDs
read_id=[]
# nb mapped reads from ERCC 
ercc_reads=[]
# nb mapped reads from mouse genome
cell_reads=[]
# total nb reads 
tot_reads=[]

# Fill mouse array with IDs and number of mapped cell reads 
with open(dir_mouse_map, 'r') as mmapped:
    countm = 0
    for line in mmapped:
        countm+=1
        if countm % 2 == 0:
            mreads=line.split('\n')
#             print('mapped reads: '+mreads[0])
            cell_reads.append(mreads[0])
        else:
            name=line.split('_')
            read_id.append(name[0])
#             print('ID: '+name[0])
    print('cnt mapped: '+str(countm))
    

# Fill array with number of total reads    
with open(dir_ERCC_tot, 'r') as etot:
    countt = 0
    for line in etot:
        countt+=1
        if countt % 2 == 0:
            treads=line.split('\n')
#             print('total reads: '+line)
            tot_reads.append(treads[0])
        else:
            name=line.split('_')
#             print('ID '+ name[0])
#             read_id_tot.append(name[0])

    
with open(dir_ERCC_map, 'r') as etot:
    countt = 0
    for line in etot:
        countt+=1
        if countt % 2 == 0:
            treads=line.split('\n')
#             print('total reads: '+line)
            ercc_reads.append(treads[0])
        else:
            continue

            
a= np.array(read_id)
b= list(map(int, ercc_reads))
c= list(map(int, cell_reads))
e= list(map(int, tot_reads))
d= np.subtract(e,np.add(b,c))
f= np.divide(b,e)
g= np.divide(c,e)
# Compute Median cell size
ercc_alj_norm=np.median(np.divide(b,np.add(b,c)))
ercc_ali_norm=np.divide(b,np.add(b,c))
mouse_alj_norm=np.median(np.divide(c,np.add(b,c)))
mouse_ali_norm=np.divide(c,np.add(b,c))

h= np.divide(ercc_alj_norm, ercc_ali_norm)    
i= np.divide(mouse_alj_norm, mouse_ali_norm) 
df=pd.DataFrame({'Cell':a, 'ERCC Reads':b, 'Cell Reads':c,'Unclassified Reads':d, 'Total Reads':e, 'ERCC Reads (Fraction)':f, 
                 'Cell Reads (Fraction)':g, 'ERCC Scale Factor':h, 'Cell Scale Factor':i},  columns=['Cell','ERCC Reads','Cell Reads', 'Unclassified Reads', 'Total Reads', 
                                                       'ERCC Reads (Fraction)', 'Cell Reads (Fraction)', 'ERCC Scale Factor','Cell Scale Factor'])
df=df.T
print(df.T)


# # Write csv file
df.to_csv("total_reads_DistalLungEpithelium.csv", header=False)
# f = open('./total_reads_DistalLungEpithelium.csv', 'a')
# writer = csv.writer(f)
# writer.writerow(read_id)
# writer.writerow(ercc_reads)
# writer.writerow(cell_reads)
# # writer.writerow(tot_reads)
# f.close()



```
