```python
import os
import pandas as pd
import numpy as np

gtf_path='/work/stud2019/mpoets/no_backup/STAR/source/ERCC92_gtf_fa/ERCC92.gtf'
fasta_path='/work/stud2019/mpoets/no_backup/STAR/source/ERCC92_gtf_fa/ERCC92.fa'
trans_path='/work/stud2019/mpoets/no_backup/STAR/source/all_featurcnts/'
dic = {}



for line in open(gtf_path, "r"):
    data=line.split()
#     gene_id.append(data[0])
#     trans_id.append(data[11][1:-2])
    dic[data[0]]=data[11][1:-2]
# print(dic)

for file in os.listdir(trans_path):
    gene_id=[]
    trans_id=[]
    length=[]
    counts=[]
    filename=file.split('.')[0]
#     print()
    with open(trans_path+file, 'r') as fcounts:
        next(fcounts)
        next(fcounts)
        for line in fcounts:
            trans_id.append(line.split()[0])
            gene_id.append(line.split()[1])
            length.append(line.split()[5])
            counts.append(line.split()[6])
    coverage=np.divide(list(map(int, counts)),list(map(int, length)))
    df=pd.DataFrame({'Transcript_ID':trans_id, 'Gene_ID':gene_id, 'Length':length, 'Counts':counts, 'Coverage':coverage}, columns=['Transcript_ID','Gene_ID', 'Length', 'Counts', 'Coverage'])
#     print(df)
    df.to_csv("./CoverageCnt/"+filename+'.csv', index=False)
```
